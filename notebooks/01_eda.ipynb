# 01 — Exploratory Data Analysis
Aurel & Co. weekly performance data. Real logic lives in `src/`; this notebook is for visual exploration only.

```python
import sys, pathlib
sys.path.insert(0, str(pathlib.Path.cwd().parent))
import pandas as pd
import plotly.express as px
from src.data.load_data import load_config, load_raw_data

cfg = load_config()
df = load_raw_data(config=cfg)
df.head()
```

## Revenue & units over time

```python
fig = px.line(df, x='week_start', y=['revenue','units_sold'])
fig.show()
```

## Spend by channel over time

```python
channel_cols = [c['spend_col'] for c in cfg['channels']]
fig = px.area(df, x='week_start', y=channel_cols)
fig.show()
```

## Naive correlation check (raw spend vs revenue, ignoring adstock)
Expect these to look weak/misleading — that's the whole reason adstock+saturation exist. Don't over-interpret this cell.

```python
df[channel_cols + ['revenue']].corr()['revenue']
```

## Promo / discount / seasonality sanity checks

```python
print(df.groupby('promo_flag')['units_sold'].mean())
print(df.groupby('stock_out_flag')['units_sold'].mean())
print(df.groupby('delivery_delay_flag')['returns_pct'].mean())
```
