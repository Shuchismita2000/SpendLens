# 03 — Model Validation
Expanding-window CV results, recovered-vs-true parameter check, holdout diagnostics. Production code: `src/models/train_model.py`, `src/models/evaluate.py`.

```
import sys, pathlib, json
sys.path.insert(0, str(pathlib.Path.cwd().parent))
import pandas as pd, plotly.express as px, plotly.graph_objects as go
from src.data.load_data import load_config

cfg = load_config()
cv_report = json.load(open(pathlib.Path.cwd().parent / cfg['paths']['cv_report']))
recovery = json.load(open(pathlib.Path.cwd().parent / cfg['paths']['recovery_report']))
cv_report['cv_metrics']
```

## Adstock recovery: fitted vs. true lambda per channel

```
rec_df = pd.DataFrame(recovery['adstock_recovery'])
fig = px.bar(rec_df, x='channel', y=['true_lambda','fitted_lambda'], barmode='group')
fig.show()
rec_df
```

## Holdout: actual vs predicted units_sold

```
diag = json.load(open(pathlib.Path.cwd().parent / 'outputs/reports/holdout_diagnostics.json'))
diag_df = pd.DataFrame(diag)
fig = go.Figure()
fig.add_trace(go.Scatter(x=diag_df['week_start'], y=diag_df['actual'], name='actual'))
fig.add_trace(go.Scatter(x=diag_df['week_start'], y=diag_df['predicted'], name='predicted'))
fig.show()
```
